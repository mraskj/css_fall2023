```python
import os

import numpy as np

from pathlib import Path
from scipy.io import wavfile

import matplotlib.pyplot as plt
import seaborn as sns

import parselmouth

sns.set() # Use seaborn's default style to make attractive graphs
```

```python
file_dir = 'C://Users//au535365//Dropbox//research//stigma-paper//clips'
```

```python
files = [x for x in os.listdir(file_dir) if x.endswith('wav')]
fpaths = [str(Path(os.path.join(file_dir, z))) for z in files]
```

```python
sr, data = wavfile.read(fpaths[0])
```

```python
dur = data.shape[0]/16000
```

```python
xs = np.linspace(3.12500000e-05, dur, data.shape[0])
```

```python
min(data)
```

```python
plt.figure()
plt.plot(xs[:int(sr*10)], data[:int(sr*10)])
plt.xlabel("time [s]")
plt.ylabel("amplitude")
plt.show()
```
